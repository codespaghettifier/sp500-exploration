```python
import pandas as pd
from yahoofinancials import YahooFinancials
from datetime import datetime
import json
```

```python
ticker = tickers[0]
yahoo_financials = YahooFinancials(ticker)
financial_statements = yahoo_financials.get_financial_stmts('quarterly', ['income', 'balance', 'cash'])
```

```python
incomes = financial_statements['incomeStatementHistoryQuarterly'][ticker]
balances = financial_statements['balanceSheetHistoryQuarterly'][ticker]
cashflows = financial_statements['cashflowStatementHistoryQuarterly'][ticker]
```

```python
def reaplace_date_with_timestamp(statements):
    for statement in statements:
        dates = list(statement.keys())
        for date in dates:
            timestamp = int(datetime.strptime(date, '%Y-%m-%d').timestamp())
            statement[timestamp] = statement[date]
            del statement[date]
```

```python
incomes_copy = incomes.copy()
reaplace_date_with_timestamp(incomes_copy)
incomes_copy
```

```python
def fetch_company_financials(ticker):
    yahoo_financials = YahooFinancials([ticker])

    quarterly_statements = yahoo_financials.get_financial_stmts('quarterly', ['income', 'balance', 'cash'])
    annual_statements = yahoo_financials.get_financial_stmts('annual', ['income', 'balance', 'cash'])

    quarterly_income = quarterly_statements['incomeStatementHistoryQuarterly'][ticker]
    quarterly_balance = quarterly_statements['balanceSheetHistoryQuarterly'][ticker]
    quarterly_cashflow = quarterly_statements['cashflowStatementHistoryQuarterly'][ticker]
    annual_income = annual_statements['incomeStatementHistory'][ticker]
    annual_balance = annual_statements['balanceSheetHistory'][ticker]
    annual_cashflow = annual_statements['cashflowStatementHistory'][ticker]

    financials = {
        'quarterly_income': quarterly_income,
        'quarterly_balance': quarterly_balance,
        'quarterly_cashflow': quarterly_cashflow,
        'annual_income': annual_income,
        'annual_balance': annual_balance,
        'annual_cashflow': annual_cashflow
    }
    
    for statements in financials.keys():
        reaplace_date_with_timestamp(financials[statements])

    return financials
```

```python
financials = fetch_company_financials('A')
financials
```

```python
def save_tickers_to_file(tickers):
    file_path = 'data/yahoo/tickers.json'
    with open(file_path, 'w') as json_file:
        json.dump(tickers, json_file, indent=4)
```

```python
def save_financials_to_file(ticker, financials):
    file_path = f'data/yahoo/financials_{ticker}.json'
    with open(file_path, 'w') as json_file:
        json.dump(financials, json_file, indent=4)
```

```python
ticker = tickers[0]
save_financials_to_file(ticker, financials)
```

```python
def save_fetch_log_to_file(log):
    file_path = 'data/yahoo/fetch_log.json'
    with open(file_path, 'w') as json_file:
        json.dump(log, json_file, indent=4)
```

```python
def fetch_and_save_financials(tickers):
    fetch_log = []
    for index, ticker in enumerate(tickers):
        print(f'[{index + 1}\t/ {len(tickers)}] Fetching {ticker} ... ', end='')
        try:
            financials = fetch_company_financials(ticker)
        except:
            fetch_log.append({'ticker': ticker, 'fetch success': False})
            print('FAILED')
            continue

        print(f'OK, saving ... ', end='')
        try:
            save_financials_to_file(ticker, financials)
        except:
            fetch_log.append({'ticker': ticker, 'fetch success': True, 'save success': False})
            print('FAILED')
            continue
        print(f'OK')

        fetch_log.append({'ticker': ticker, 'fetch success': True, 'save success': True})
    
    save_fetch_log_to_file(fetch_log)
    return fetch_log
```

```python
fetch_log = fetch_and_save_financials(tickers[:2])
```

```python
fetch_log
```

```python
def fetch_log_to_fetch_success(fetch_log):
    fetch_success = {}
    for log in fetch_log:
        fetch_success[log['ticker']] = log['fetch success'] and log['save success']
    return fetch_success
```

```python
fetch_success = fetch_log_to_fetch_success(fetch_log)
fetch_success
```

```python
def save_merged_financials_to_file(merged_financials):
    file_path = f'data/yahoo/merged_financials.json'
    with open(file_path, 'w') as json_file:
        json.dump(merged_financials, json_file, indent=4)
```

```python
def merge_financials(tickers):
    merged_financials = {}
    for ticker in tickers:
        file_path = f'data/yahoo/financials_{ticker}.json'
        with open(file_path, 'r') as json_file:
            financials = json.load(json_file)

        merged_financials[ticker] = financials
    return merged_financials
```

```python
successfully_fetched_tickers = [ticker for ticker, success in fetch_success.items() if success]
merged_financials = merge_financials(successfully_fetched_tickers)
save_merged_financials_to_file(merged_financials)
```

# Final fetching and saving

```python
sp_500_value_df = pd.read_csv('data/data_origin/FS_sp500_Value.csv')
tickers = pd.Series(sp_500_value_df['Ticker'].unique()).to_list()

fetch_log = fetch_and_save_financials(tickers[:10])

fetch_success = fetch_log_to_fetch_success(fetch_log)
successfully_fetched_tickers = [ticker for ticker, success in fetch_success.items() if success]

merged_financials = merge_financials(successfully_fetched_tickers)
save_merged_financials_to_file(merged_financials)
```
